5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
convert = [
    {
        "$addFields": {
            "geocode.longitude": { "$toDouble": "$geocode.longitude" },
            "geocode.latitude": { "$toDouble": "$geocode.latitude" }
        }
    }
]

# Perform the aggregation
result = establishments.aggregate(convert)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
convert = [
    {
        "$addFields": {
            "RatingValue": { "$toInt": "$RatingValue" }
        }
    }
]

# Perform the aggregation
result = establishments.aggregate(convert)
```

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Query to find documents with non-1-5 rating values
query = {"Rating": {"$not": {"$in": [1, 2, 3, 4, 5]}}}

# Retrieve the documents that match the query
documents = establishments.find(query)

# Update each document individually
updated_count = 0
for document in documents:
    updated_document = {"$set": {"Rating": None}}
    establishments.update_one({"_id": document["_id"]}, updated_document)
    updated_count += 1
```

```python
# Change the data type from String to Integer for RatingValue
update_operation = [
    {
        "$set": {
            "RatingValue": {
                "$toInt": "$RatingValue"
            }
        }
    }
]

# Perform the update
establishments.update_many({}, update_operation)
```

```python
# Check that the coordinates and rating value are now numbers
for doc in documents:
    longitude = doc['geocode']['longitude']
    latitude = doc['geocode']['latitude']
    rating = doc['RatingValue']
    print(f"Longitude: {longitude}, Latitude: {latitude}, Rating: {rating}")
```
